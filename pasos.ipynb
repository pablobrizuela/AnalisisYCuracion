```python
import pandas as pd
import seaborn
import matplotlib.pyplot as plt
```

```python
pasos = pd.read_json("pasos.json")
```

```python
pasos_por_provincia = pasos.groupby('province').size()
pasos_por_provincia
```

```python
plt.figure(figsize=(10,6))
seaborn.barplot(x=pasos_por_provincia.keys(),y=pasos_por_provincia)
plt.xticks(rotation=90)
```
